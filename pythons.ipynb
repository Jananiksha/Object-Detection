```python
import io
import cv2
import requests
import numpy as np
from PIL import Image
from requests_toolbelt.multipart.encoder import MultipartEncoder

# Load the image using OpenCV
img = cv2.imread('final.jpg')  # Replace with your image file path
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB
pilImage = Image.fromarray(image)

# Convert the image to a JPEG buffer
buffered = io.BytesIO()
pilImage.save(buffered, format="JPEG", quality=100)
buffered.seek(0)  # Reset buffer pointer to the beginning

# Define the multipart data
m = MultipartEncoder(
    fields={'file': ('imageToUpload', buffered.getvalue(), 'image/jpeg')}
)

# Make the POST request
response = requests.post(
    "https://detect.roboflow.com/safetyobjectdetection/1?api_key=AB2PH7hlm0Wcfyii4x72",
    data=m,
    headers={'Content-Type': m.content_type}
)

# Print the response
print(f"Status Code: {response.status_code}")
try:
    print(f"Response: {response.json()}")
except Exception as e:
    print(f"Error decoding response as JSON: {e}")
    print(response.text)
```

local image using base64

```python
import requests
import base64
import io
from PIL import Image

#Load Image with PIL
image = Image.open("final.jpg").convert("RGB")

#Convert to JPEG Buffer
buffered = io.BytesIO()
image.save(buffered,quality=100, format='JPEG')

#Encode to base64
img_str = base64.b64encode(buffered.getvalue())
img_str = img_str.decode('ascii')

#Construct the URL
upload_url = "".join([
    "https://detect.roboflow.com/safetyobjectdetection/1"
    "?api_key=AB2PH7hlm0Wcfyii4x72"
    "&name=YOUR_IMAGE.jpg"
])

# POST to the API
r = requests.post(upload_url, data=img_str, headers={"Content-Type": "application/x-www-form-urlencoded"})
print(r.json())

```

Image to base64 conversion

```python
import base64

# Path to your image file
image_path = 'final.jpg'  # Replace with the path to your image

# Open the image in binary mode
with open(image_path, "rb") as image_file:
    # Convert the image to a Base64 string
    base64_string = base64.b64encode(image_file.read()).decode('utf-8')

# Print the Base64 string
print(base64_string)
```

Image Encoding for base64

```python
from PIL import Image
import base64
import io

# Open the image
image = Image.open('final.jpg')

# Compress the image (reduce quality)
compressed_image = io.BytesIO()
image.save(compressed_image, format="JPEG", quality=50)  # Adjust quality (lower = smaller size)
compressed_image.seek(0)

# Convert compressed image to Base64
base64_string = base64.b64encode(compressed_image.read()).decode('utf-8')

# Print the shorter Base64 string
print(base64_string)
```

```python
import hashlib

# Path to your image file
image_path = 'final.jpg'

# Read the image in binary mode
with open(image_path, "rb") as image_file:
    image_data = image_file.read()

# Generate a hash of the image
image_hash = hashlib.sha256(image_data).hexdigest()

# Print the hash
print(f"Image Hash: {image_hash}")
```
